```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
%matplotlib inline
```

```python
# @@@@  assignment @@@@@
iris=sns.load_dataset('iris')
iris.head()
```

```python
a=iris.iloc[:,:-1].values
b=iris.iloc[:,4].values
```

```python
# label encoding the species(categorical) values

from sklearn.preprocessing import LabelEncoder
labelencoder_b=LabelEncoder()
b1=labelencoder_b.fit_transform(b)
```

```python
# feature scaling

from sklearn.preprocessing import StandardScaler
stdscaler_a=StandardScaler()
a1=stdscaler_a.fit_transform(a)
```

```python
# reduction of dimension from 4 to 3d

from sklearn.decomposition import PCA
pca=PCA(n_components=3)
a2=pca.fit_transform(a1)
```

```python
# visualizing the 3D chart

from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure(1,figsize=(8,5))
ax=Axes3D(fig,rect=[0,0, 1, 1],elev=-150, azim=110)

ax.scatter(a2[:,0],a2[:,1],a2[:,2],c=b1,edgecolor='k', s=30,cmap=plt.cm.rainbow_r)
    
ax.set_xlabel("PCA - 1")
ax.set_ylabel("PCA - 2")
ax.set_zlabel("PCA - 3")

plt.show()
```

```python
iris = datasets.load_iris() 
X_reduced = PCA(n_components=3).fit_transform(iris.data) 
Y = iris.target

fig = plt.figure(1, figsize=(8, 6)) 
ax = Axes3D(fig, elev=-150, azim=110)

ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=Y, cmap=plt.cm.rainbow_r)

ax.set_title("First three PCA directions") 
ax.set_xlabel("1st eigenvector") 
ax.set_ylabel("2nd eigenvector") 
ax.set_zlabel("3rd eigenvector")

plt.show()
```

```python
################ assignment1 ####################

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn.decomposition import PCA
from sklearn import datasets

iris=datasets.load_iris()
a,b = iris.data , iris.target

pca=PCA(n_components=3)
a1=pca.fit_transform(a)

iris_fig=plt.figure(1,figsize=(7,6))
iris_plot=Axes3D(iris_fig,rect=[0,0, 1, 1],elev=-140, azim=120)

iris_plot.scatter(a1[:,0],a1[:,1],a1[:,2],c=b,edgecolor='k', s=35,cmap=plt.cm.rainbow)
    
iris_plot.set_xlabel("PCA - 1")
iris_plot.set_ylabel("PCA - 2")
iris_plot.set_zlabel("PCA - 3")
iris_plot.set_title("PCA Three Components Visualization")
plt.show()
```

```python
import numpy as np
from scipy import optimize

Max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
Min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
months=np.arange(12)
plt.figure() 
plt.plot(months,Min, 'bo') 
plt.plot(months,Max, 'ro') 
plt.show()

def temp_of_year(t, ini, A, t_off): 
    return (ini + A * np.cos((t + t_off) * 1.8 * np.pi / t.max()))

r_max, c_max = optimize.curve_fit(temp_of_year, months, Max, [60, 30, 0]) 
r_min, c_min = optimize.curve_fit(temp_of_year, months, Min, [-60, 30, 0]) 
#print(r_max,r_min)

```

```python
d = np.linspace(0, 12) 
plt.figure() 
plt.plot(months,Max, 'ro')  
plt.plot(months,Min, 'bo') 
plt.plot(d, temp_of_year(d, *r_max), 'r-')
plt.plot(d, temp_of_year(d, *r_min), 'b-') 
plt.title('Temperatures for each months of the year')
plt.xlabel('Month') 
plt.ylabel('Temperature (°C)') 
plt.show()
```

```python
#defining periodic function 
from scipy import optimize
def yearly_temps(times, avg, ampl, time_offset): 
    return (avg + ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))

#identifying the curve for plot using scipy 
res_max, cov_max = optimize.curve_fit(yearly_temps, months,Max, [40, 20, 0]) 
res_min, cov_min = optimize.curve_fit(yearly_temps, months,Min, [-40, 20, 0]) 
print(res_max) 
print(res_min) 
#[ 39.10062829 -11.0872493 2.10332481] [24.87112259 -7.89899664 1.60673368]
#2. plot the fit

days = np.linspace(0, 12, num=365) 
plt.figure() 
#plt.plot(months,Max, 'ro') 
plt.plot(days, yearly_temps(days, *res_max), 'r-') 
#plt.plot(months,Min, 'bo') 
plt.plot(days, yearly_temps(days, *res_min), 'b-') 
plt.xlabel('Month') 
plt.ylabel('Temperature (°C)') 
plt.show()
```
