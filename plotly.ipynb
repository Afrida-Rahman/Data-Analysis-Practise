```python
df=pd.DataFrame(randn(10,3),columns=['a','b','c'])
df['a']=df['a'].cumsum()+20
df['b']=df['b'].cumsum()+20
df['c']=df['c'].cumsum()+20
df.head()
```

```python
df.iplot()
```

```python
df.iplot(x='a',y='b',mode='markers',size=25)
```

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.iplot(kind='bar',x='sex',y='survived',title='Survived person',xTitle='Sex',yTitle='Survived')
```

```python
# theme
cf.getThemes()
```

```python
cf.set_config_file(theme='henanigans')
titanic.iplot(kind='bar',x='sex',y='survived',title='Survived person',xTitle='Sex',yTitle='Survived')
```

```python
df.iplot(kind='bar',barmode='stack',bargap=0.5)
```

```python
# horixontal bargraph
df.iplot(kind='barh',barmode='stack',bargap=0.5)
```

```python
# box plot
df.head()
```

```python
# IQR=inter quartile range(q3-q1)
# max=q3 + 1.5 * IQR
# min=q1 -1.5 * IQR
# any data below min/above max=outlier
# q3=75th
# q2=50th
# q1=25th

df.iplot(kind='box')
```

```python
df.iplot(kind='area')
```
